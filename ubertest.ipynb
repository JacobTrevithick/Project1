```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress

uber_raw_df = pd.read_csv('uberCSV/uber-raw-data-apr14.csv')

weather_df = pd.read_csv('Data/NewYorkWeatherData_2014_2015.csv')
```
